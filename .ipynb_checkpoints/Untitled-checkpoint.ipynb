```python
import numpy as np
import librosa
import math
from playsound import playsound
import matplotlib.pyplot as plt
import librosa.display
```

```python
# test .wav can playsound
filename = '2.wav'
playsound(filename)
y, sr = librosa.load(filename, sr=None)
```

```python
plt.figure(figsize=(25, 5))
librosa.display.waveplot(y, sr=sr)
plt.title('waveform')
plt.show()
```

```python
# w(n) = 1/(2*N) if 0~N-1 else 0
def sgn(arr):
    for i in range(len(arr)):
        if(arr[i]>=0):
            arr[i] = 1
        else:
            arr[i] = -1
    return arr
def calZcr(waveData,frameSize,overLap):
    wlen = len(waveData)
    step = frameSize - overLap
    frameNum = math.ceil(wlen/step)
    zcr = np.zeros((frameNum,1))
    for i in range(frameNum):
        curFrame = waveData[np.arange(i*step,min(i*step+frameSize,wlen))]
        curFrame = sgn(curFrame)
#         zcr[i] = sum(curFrame[:-1]*curFrame[1:]<=0)
        zcr[i] = sum(abs(curFrame[1:] - curFrame[:-1]))/(2*frameSize)
    return zcr

# print(len(y))
frameSize = 0.02/sr
overLap = 0
zcr = calZcr(y,frameSize,overLap)

# print(len(zcr))
time_zcr = np.arange(0, len(zcr)) * (len(y)/(len(zcr)*sr))
plt.figure(figsize=(25, 5))
plt.plot(time_zcr, zcr)
plt.ylabel("ZCR")
plt.xlabel("time (s)")
plt.title('Zero-crossing rate contour')
plt.show()
```

```python
# w(n) = rectangular window 1 if 0~N-1 else 0
def calEnergy(waveData,frameSize,overLap):
    wlen = len(waveData)
    step = frameSize - overLap
    frameNum = math.ceil(wlen/step)
    energy = np.zeros((frameNum,1))
    for i in range(frameNum):
        curFrame = waveData[np.arange(i*step,min(i*step+frameSize,wlen))]
        energy[i] = sum(curFrame**2)
    return energy

frameSize = 0.02/sr
overLap = 0
energy = calEnergy(y,frameSize,overLap)

# print(len(zcr))
time_energy = np.arange(0, len(energy)) * (len(y)/(len(energy)*sr))
plt.figure(figsize=(25, 5))
plt.plot(time_energy, energy)
plt.ylabel("Energy")
plt.xlabel("time (s)")
plt.title('Energy contour')
plt.show()
```

```python
# w(n) = 1/(N) if 0~N-1 else 0
def calMagnitude(waveData,frameSize,overLap):
    wlen = len(waveData)
    step = frameSize - overLap
    frameNum = math.ceil(wlen/step)
    magnitude = np.zeros((frameNum,1))
    for i in range(frameNum):
        curFrame = waveData[np.arange(i*step,min(i*step+frameSize,wlen))]
        magnitude[i] = sum(abs(curFrame))/(frameSize)
    return magnitude

frameSize = 0.02/sr
overLap = 0
magnitude = calMagnitude(y,frameSize,overLap)
time_magnitude = np.arange(0, len(magnitude)) * (len(y)/(len(magnitude)*sr))
plt.figure(figsize=(25, 5))
plt.plot(time_magnitude, magnitude)
plt.ylabel("magnitude")
plt.xlabel("time (s)")
plt.title('Average Magnitude')
plt.show()
```

```python
# center clip
CL = 0.3*np.max(abs(y.copy()))
x = np.zeros((len(y),1))
for i in range(len(y)):
    if y[i]>=CL:
        x[i] = y[i]-CL
    elif abs(y[i])<CL:
        x[i] = 0
    else:
        x[i] = y[i]+CL
time_center_clip = np.arange(0,len(x))/sr
plt.figure(figsize=(25, 5))
plt.plot(time_center_clip, x)
plt.ylabel("center clip speech")
plt.xlabel("time (s)")
plt.title('center clip')
plt.show()

# calculate Rn(k) use autocorrelation
def calPitch(waveData,frameSize,overLap):
    R = 0
    Rn = 0
    maxk = 0
    step = frameSize - overLap
    wlen = len(waveData)
    frameNum = math.ceil(wlen/step)
    R = np.zeros((frameNum,1))
    for i in range(frameNum):
        for k in range(step):
            Rn = np.zeros((step,1))
            if i*step+frameSize+k<wlen:
                curFrame = x[np.arange(i*step,min(i*step+frameSize,wlen))]
                nxtFrame = x[np.arange(i*step+k,min(i*step+frameSize+k,wlen))]
                Rn[k] = sum(curFrame*nxtFrame)
                if abs(Rn[k])>abs(Rn[maxk]):
                    maxk = k
        R[i] = maxk
        maxk = 0
    return R

frameSize = 0.02/sr
overLap = 0
pitch = calPitch(x,frameSize,overLap)
time_pitch = np.arange(0, len(pitch)) * (len(x)/(len(pitch)*sr))
plt.figure(figsize=(25, 5))
plt.plot(time_pitch, pitch)
plt.ylabel("pitch")
plt.xlabel("time (s)")
plt.title('pitch detection')
plt.show()
```
